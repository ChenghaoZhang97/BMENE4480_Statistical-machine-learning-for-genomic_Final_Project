```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
f1_improved = [0, 0, 0, 0.984, 0.964, 0.994, 0.96875]
f1_gini = [0, 0, 0, 0, 0, 0, 0.946]
proportion = [1, 2, 3, 5, 7, 9, 10]
```

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(proportion, f1_improved, '-o', label='ImprovedGiniClust')
plt.plot(proportion, f1_gini, '-o', label='GiniClust3')
plt.title("F1 score")
plt.xlabel("Percentage of rare cells (%)")
plt.ylabel("F1 score")
plt.legend()
plt.show()
```
